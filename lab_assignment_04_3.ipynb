```python
import pandas #pd
url = 'https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab-assignment-04-student-performance.csv'
# reading data from git
dt = pd.read_csv(url)
dt.head()

#reading the csv file to dataframe and iterating through each column to get the number of rows,number of unique values
for cl in dt.columns:
    print("Column:",cl)
    print("The no of rows are:", dt[cl].shape[0])
    print("The no of unique values in this column are:", dt[cl].value_counts().shape[0])
    print(dt[cl].value_counts())
    print()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.countplot(dt['sh'])
plt.xlabel("Sh")
plt.ylabel("C")
plt.show()
```

```python
import random
import seaborn as sns
import pandas as pd
d_new = pd.DataFrame(dt.sh)
d_new['esp'] = pd.DataFrame(dt.esp)
for ind in dt.index:
    if d_new['esp'][ind] == "Best":
      d_new['esp'][ind] = random.uniform(80,100)
    elif d_new['esp'][ind] == "Good":
      d_new['esp'][ind] = random.uniform(60,79)
    elif d_new['esp'][ind] == "Vg":
      d_new['esp'][ind] = random.uniform(45,59)
    elif d_new['esp'][ind] == "Pass":
      d_new['esp'][ind] = random.uniform(30,44)
    elif d_new['esp'][ind] == "Fail":
      d_new['esp'][ind] = random.uniform(0,29)
sns.factorplot(x='sh',y='esp', kind = 'strip',data=d_new)
```

```python
#Import label encoder 
from sklearn import preprocessing
import seaborn as sns
l_encoder = preprocessing.LabelEncoder()  
def Encoder(df):
    columnsToEncode = list(df.select_dtypes(include=['category','object']))
    le = preprocessing.LabelEncoder()
    # we convert categorical columns data into numerical data
    for feature in columnsToEncode:
        try:
            df[feature] = le.fit_transform(df[feature])
        except:
            print('Error encoding '+feature)
    return df

dt1 = Encoder(dt)
dt1 = dt1.drop('ms', axis = 1)
corr = dt1.corr()
sns.set(rc = {'figure.figsize':(50,30)})
sns.heatmap(corr,annot=True)
```
